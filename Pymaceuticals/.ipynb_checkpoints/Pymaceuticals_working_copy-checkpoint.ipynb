## Observations and Insights 

```python
# Observations:
# 1. Looking at the timepoint is critical when evaluating the effectiveness
# of a drug, as a drug may appear to be quickly diminishing the tumor size early on, 
# only to have in increase rapidly later on. 

# 2. There appears to be a connection between tumor size and weight, with
# tumor size increasing with weight. 

# 3. At first glance, Capomulin and Ramicane appear to be the most effective.
# However, it would be interesting to study these drugs over a longer period of time 
# to see if the upward trend with tumor size in the Capomulin study continues
# and for how long (i.e., could you possibly end up with a larger tumor than
# you started with if that growth continues over an extended period?)
```

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as stats
import numpy as np

# Import study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results using Pandas
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Merge the data into a single dataset using Pandas
combined_df=pd.merge(mouse_metadata, study_results)

# Display the first 10 rows of the data table for preview
combined_df.head(10)
```

## Summary Statistics

```python
# Use groupby and summary statistical methods to calculate the following 
#properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 

# Assemble the resulting series into a single summary dataframe.
```

```python
# List dataframe column names
study_mice.columns
```

```python
regimen_grouped=study_mice.groupby("Drug Regimen")

mean=regimen_grouped.mean()["Tumor Volume (mm3)"]
# mean

median=regimen_grouped.median()["Tumor Volume (mm3)"]
# median

variance=regimen_grouped.var()["Tumor Volume (mm3)"]
# variance

standard_dev=regimen_grouped.std()["Tumor Volume (mm3)"]
# standard_dev

sem=regimen_grouped.sem()["Tumor Volume (mm3)"]
# sem
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
summary_data_df=pd.DataFrame({
    "Mean": mean,
    "Median": median,
    "Variance": variance,
    "Standard Deviation": standard_dev,
    "SEM": sem
})
summary_data_df
```

```python
# Using the aggregation method, produce the same summary statistics in a single line
aggregate_data=regimen_grouped.agg({'Tumor Volume (mm3)': ['mean', 'median', 'var', 'std', 'sem']})
aggregate_data
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
# Start by getting the last (greatest) timepoint for each unique mouse
grouped_mice=study_mice.groupby("Mouse ID")["Timepoint"].max()

# This is a series, want to find a way to only grab these rows from the main dataset 
# this holds grouped mice constant and then merges other dataframe onto it; by default does an inner merge 
# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
max_timepoint=pd.merge(grouped_mice, study_mice, on=["Mouse ID", "Timepoint"], how="left")
max_timepoint
```

```python
# Put treatments into a list for for loop (and later for plot labels)
drugs= ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]
tumor_vol_data=[]

for drug in drugs:
    drug_filter=study_mice.loc[:, "Drug Regimen"]== drug
    drug_filter=study_mice.loc[drug_filter, "Tumor Volume (mm3)"]
    drug_quantiles=drug_filter.quantile([.25,.5,.75])
    drug_quantiles
    lowerq=drug_quantiles[0.25]
    upperq=drug_quantiles[0.75]
    iqr=upperq-lowerq
    lower_bound=lowerq-(1.5*iqr)
    upper_bound=upperq-(1.5*iqr)
    tumor_vol_data=[drug_quantiles]
    print(f"Values below {lower_bound} could be outliers for {drug}.")
    print(f"Values above {upper_bound} could be outliers for {drug}.")
   # print(tumor_vol_data)
```

```python
plt.boxplot(tumor_vol_data)
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin
# Need to isolate the tumor vol and time points for specific Mouse ID: i738
# Time goes on x axis, tumor volume on y 
mouse_i738= study_mice[study_mice["Mouse ID"]=="i738"]
mouse_i738

time=mouse_i738["Timepoint"]
tumor_volume=mouse_i738["Tumor Volume (mm3)"]
plt.xlabel("Time")
plt.ylabel("Tumor Volume (mm3)")
i738_plot = plt.plot(time, tumor_volume, label = "Mouse i738")
plt.xlabel('Timepoint')
plt.ylabel('Tumor Volume (mm3)')
plt.title("Tumor Volume vs Timepoint for Capomulin Regimen - Mouse i738")
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
# Need to groupby drug regimen and isolate for Capomulin 
drug_regimen=study_mice[study_mice["Drug Regimen"]=="Capomulin"]
# drug_regimen - Returns a dataframe containing only mice treated with Capomulin 

# X axis needs to be weight (independent variable) and y axis needs to be tumor volume (dependent variable)
avg_tumor_vol=drug_regimen.groupby("Mouse ID")["Tumor Volume (mm3)"].mean()
avg_weight=drug_regimen.groupby("Mouse ID")["Weight (g)"].mean()
#labels=("Weight", "Tumor Volume")
plt.scatter(avg_weight, avg_tumor_vol)
plt.xlabel('Weight (g)')
plt.ylabel('Tumor Volume (mm3)')
plt.title("Avg Tumor Size vs Weight for Mice on Capomulin Regimen")
```

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
```

```python
drug_regimen
```

```python
# Calculate Pearsons Correlation coefficient
weight=drug_regimen.iloc[:, 4]
avg_tumor_vol=drug_regimen.iloc[:, 6]
correlation=stats.pearsonr(weight, avg_tumor_vol)
print(f"The correlation coefficient between mouse weight and average tumor volume in the Capomulin regime is {round(correlation[0],2)}")
```

```python
avg_tumor_vol=drug_regimen.groupby("Mouse ID")["Tumor Volume (mm3)"].mean()
avg_weight=drug_regimen.groupby("Mouse ID")["Weight (g)"].mean()
plt.scatter(avg_weight, avg_tumor_vol)

# Perform a linear regression on weight vs tumor volume
vc_slope, vc_int, vc_r, vc_p, vc_std_err = stats.linregress(avg_weight, avg_tumor_vol)

# Create equation of line to calculate linear regression for mice on Capomulin regimen
vc_fit = vc_slope * weight + vc_int
plt.plot(weight, vc_fit, "--")
plt.xlabel('Weight (g)')
plt.ylabel('Tumor Volume (mm3)')
plt.title("Linear Regression for Capomulin Regimen")
```
